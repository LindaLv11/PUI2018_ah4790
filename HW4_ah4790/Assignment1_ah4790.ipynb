```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Normal Distribution

```python
# Generate Data
np.random.seed(42)
normal_samples = {}
pop_mean = 50
for i in range(0,100):
    sample = []
    sample_size = np.random.randint(10,2000)
    for j in range(0,sample_size):
        sample.append(np.random.normal(pop_mean))
    normal_samples[sample_size] = sample  
```

```python
# Calculate sample means
sample_sizes = []
sample_means = []
for key in normal_samples:
    sample_means.append(np.mean(normal_samples[key]))
    sample_sizes.append(key)
```

```python
plt.figure(figsize=(6,7))
plt.scatter(sample_sizes, sample_means)
plt.axhline(pop_mean, color='red')
plt.xlabel('Sample Size')
plt.ylabel('Sample Mean')
plt.title('Scatterplot of Sample Mean Against Sample Size\n(Normal Distribution)')
plt.figtext(0.5, -0.05, 'Sample Mean Against Sample Size with Population Mean of 50 in red\nAs the sample size increases, the sample means converge to the population mean', horizontalalignment='center', fontsize=16)
```

```python
plt.hist(sample_means, bins=20)
plt.title('Histogram of Sample Means\n(Normal Distribution)')
plt.xlabel('Sample Mean')
plt.ylabel('Counts')
```

## Binomial Distribution

```python
# Generate Data
np.random.seed(42)
binomial_samples = {}
pop_mean = 50
for i in range(0,100):
    sample = []
    sample_size = np.random.randint(51,2000)
    for j in range(0,sample_size):
        sample.append(np.random.binomial(sample_size, pop_mean/sample_size))
    binomial_samples[sample_size] = sample  
```

```python
# Calculate sample means
sample_sizes = []
sample_means = []
for key in binomial_samples:
    sample_means.append(np.sum(binomial_samples[key])/len(binomial_samples[key]))
    sample_sizes.append(key)
```

```python
plt.figure(figsize=(6,7))
plt.scatter(sample_sizes, sample_means)
plt.axhline(pop_mean, color='red')
plt.xlabel('Sample Size')
plt.ylabel('Sample Mean')
plt.title('Scatterplot of Sample Mean Against Sample Size\n(Binomial Distribution)')
plt.figtext(0.5, -0.05, 'Sample Mean Against Sample Size with Population Mean of 50 in red\nAs the sample size increases, the sample means converge to the population mean but more slowly than the other distributions', horizontalalignment='center', fontsize=16)
```

```python
plt.hist(sample_means, bins=20)
plt.title('Histogram of Sample Means\n(Binomial Distribution)')
plt.xlabel('Sample Mean')
plt.ylabel('Counts')
```

## Chi-Squared Distribution

```python
# Generate Data
np.random.seed(42)
chisq_samples = {}
pop_mean = 50
for i in range(0,100):
    sample = []
    sample_size = np.random.randint(10,2000)
    for j in range(0,sample_size):
        sample.append(np.random.chisquare(pop_mean))
    chisq_samples[sample_size] = sample  
```

```python
# Calculate sample means
sample_sizes = []
sample_means = []
for key in chisq_samples:
    sample_means.append(np.mean(chisq_samples[key]))
    sample_sizes.append(key)
```

```python
plt.figure(figsize=(6,7))
plt.scatter(sample_sizes, sample_means)
plt.axhline(pop_mean, color='red')
plt.xlabel('Sample Size')
plt.ylabel('Sample Mean')
plt.title('Scatterplot of Sample Mean Against Sample Size\n(Chi-Squared Distribution)')
plt.figtext(0.5, -0.05, 'Sample Mean Against Sample Size with Population Mean of 50 in red\nAs the sample size increases, the sample means converge to the population mean, but as slow as the binomial distribution', horizontalalignment='center', fontsize=16)
```

```python
plt.hist(sample_means, bins=20)
plt.title('Histogram of Sample Means\n(Chi-Squared Distribution)')
plt.xlabel('Sample Mean')
plt.ylabel('Counts')
```
